## Calculating second-order gradient

$$
a_i = \sum_j w_{ij} x_j, 
\quad 
x_i = f(a_i)
$$

$$
\frac{\partial^2 E}{\partial w_{ij}^2}
= \ ?
$$

#### Chain rule

$$
\frac{\partial E}{\partial w_{ij}}
=
\frac{\partial E}{\partial a_i}
\frac{\partial a_i}{\partial w_{ij}}
$$

$$
\frac{\partial^2 E}{\partial w_{ij}^2}
=
\frac{\partial}{\partial w_{ij}}
\left(
\frac{\partial E}{\partial a_i}
\frac{\partial a_i}{\partial w_{ij}}
\right)
$$

#### Recall the product rule
$$
\frac{d}{dx}[u(x)v(x)] = u'(x)v(x) + u(x)v'(x)
$$

#### Apply the product rule
$$
u = \frac{\partial E}{\partial a_i}, 
\quad 
v = \frac{\partial a_i}{\partial w_{ij}}.
$$

$$
\frac{\partial^2 E}{\partial w_{ij}^2}
= \frac{\partial}{\partial w_{ij}}(u v)
= \frac{\partial u}{\partial w_{ij}}\,v
+ u\,\frac{\partial v}{\partial w_{ij}}.
$$

$$
\frac{\partial u}{\partial w_{ij}}
= \frac{\partial^2 E}{\partial a_i^2}
  \frac{\partial a_i}{\partial w_{ij}},
\quad
\frac{\partial v}{\partial w_{ij}}
= \frac{\partial^2 a_i}{\partial w_{ij}^2}.
$$

#### Substitute $u$ and $v$ back
$$
\frac{\partial^2 E}{\partial w_{ij}^2}
= 
\left(\frac{\partial^2 E}{\partial a_i^2}
      \frac{\partial a_i}{\partial w_{ij}}\right)
\frac{\partial a_i}{\partial w_{ij}}
+
\frac{\partial E}{\partial a_i}
\frac{\partial^2 a_i}{\partial w_{ij}^2}.
$$

#### Plugging in knowns
$$
\frac{\partial a_i}{\partial w_{ij}} = x_j
\quad
\text{and}
\quad
\frac{\partial^2 a_i}{\partial w_{ij}^2} = 0.
$$

$$
\frac{\partial^2 E}{\partial w_{ij}^2}
= 
\left(\frac{\partial^2 E}{\partial a_i^2}
      x_j\right)
x_j
+
\cancel{\frac{\partial E}{\partial a_i}
\frac{\partial^2 a_i}{\partial w_{ij}^2}}.
$$


#### The second derivative term in estimate of $\Delta E$ can be calculated with
$$
\boxed{
\frac{\partial^2 E}{\partial w_{ij}^2}
= \frac{\partial^2 E}{\partial a_i^2} \, x_j^2
}
$$

$$
\frac{\partial^2 E}{\partial a_i^2} = \ ?
$$

#### Recall first-order gradient backpropagation

$$
a_i = \sum_j w_{ij} x_j, 
\quad 
x_i = f(a_i)
$$

$$
\frac{\partial E}{\partial w_{ij}} = \ ?
$$

$$
\frac{\partial E}{\partial w_{ij}}
= \frac{\partial E}{\partial a_i}
  \frac{\partial a_i}{\partial w_{ij}}
= \frac{\partial E}{\partial a_i} \, x_j
$$

$$
\frac{\partial E}{\partial a_i} = \ ?
$$

$$
\frac{\partial E}{\partial a_i}
= \frac{\partial E}{\partial x_i}
  \frac{\partial x_i}{\partial a_i}
= \frac{\partial E}{\partial x_i} \, f'(a_i)
$$

$$
\frac{\partial E}{\partial x_j} = \ ?
$$

$$
\frac{\partial E}{\partial x_j}
= \sum_k \frac{\partial E}{\partial a_k}
  \frac{\partial a_k}{\partial x_j}
$$

$$
\frac{\partial a_k}{\partial x_j} = w_{k j}
$$

$$
\frac{\partial E}{\partial x_j}
= \sum_k \left(\frac{\partial E}{\partial a_k}\right)
  w_{k j}
$$

$$
\frac{\partial E}{\partial x_j}
= \sum_k \left( \frac{\partial E}{\partial x_k} f'(a_k) \right) w_{k j}
$$

$$
\boxed{
\frac{\partial E}{\partial x_j}
= \sum_k \frac{\partial E}{\partial x_k} \, f'(a_k) \, w_{k j}
}
$$

$$
\boxed{
\frac{\partial E}{\partial a_i}
= \frac{\partial E}{\partial x_i} \, f'(a_i)
}
$$

$$
\boxed{
\frac{\partial E}{\partial w_{ij}}
= \frac{\partial E}{\partial a_i} \, x_j
}
$$


#### Second-order gradient backpropagation
$$
\frac{\partial^2 E}{\partial w_{ij}^2}
= \frac{\partial^2 E}{\partial a_i^2} \, x_j^2
$$

$$
\frac{\partial^2 E}{\partial a_i^2} = \ ?
$$

Chain rule

$$
\frac{\partial^2 E}{\partial a_i^2} = 
\frac{\partial}{\partial a_i}\left(\frac{\partial E}{\partial a_i}\right)
= \frac{\partial}{\partial a_i}\left(\frac{\partial E}{\partial x_i}
  \frac{\partial x_i}{\partial a_i}\right)
$$

$$
\frac{\partial^2 E}{\partial a_i^2}
=
\frac{\partial}{\partial a_i}
\!\left[
  \left(
    \frac{\partial E}{\partial x_i}
  \right)
  \left(
    \frac{\partial x_i}{\partial a_i}
  \right)
\right]
=
\left(
  \frac{\partial}{\partial a_i}
  \frac{\partial E}{\partial x_i}
\right)
\left(
  \frac{\partial x_i}{\partial a_i}
\right)
\;+\;
\left(
  \frac{\partial E}{\partial x_i}
\right)
\left(
  \frac{\partial}{\partial a_i}
  \frac{\partial x_i}{\partial a_i}
\right)
$$

$$
=
\left(
  \frac{\partial}{\partial a_i}
  \frac{\partial E}{\partial x_i}
\right)
f'(a_i)
\;+\;
\left(
  \frac{\partial E}{\partial x_i}
\right)
f''(a_i)
$$

$$
  \frac{\partial}{\partial a_i}
  \frac{\partial E}{\partial x_i} = \ ?
$$

$$
\frac{\partial}{\partial a_i}\frac{\partial E}{\partial x_j}
= \frac{\partial}{\partial a_i}\left(\sum_k \frac{\partial E}{\partial x_k} \, f'(a_k) \, w_{kj}\right)
$$

$$
= \sum_{k}\frac{\partial}{\partial a_i}\!\left(\frac{\partial E}{\partial x_k} \, f'(a_k) \, w_{k j}\right)
$$

Product rule
$$
= \sum_{k}\left[
\left(
\frac{\partial}{\partial a_i}
\frac{\partial E}{\partial x_k}
\right)
\left(f'(a_k)\,w_{k j}\right)
\;+\;
\left(\frac{\partial E}{\partial x_k}\right)
\left(
\frac{\partial}{\partial a_i}\left(f'(a_k)\,w_{k j}\right)
\right)
\right]
$$

$$
= \sum_{k}\left[
\left(
\frac{\partial}{\partial a_i}
\frac{\partial E}{\partial x_k}
\right)
\left(f'(a_k)\,w_{k j}\right)
\;+\;
\left(\frac{\partial E}{\partial x_k}\right)
\left(
f''(a_k)\,\frac{\partial a_k}{\partial a_i}\,w_{k j}
\right)
\right]
$$

$$
\frac{\partial a_k}{\partial a_i} = \ ?
$$

$$
x_i = f(a_i)
\quad\text{and}\quad
a_k = \sum_i w_{k i}\, x_i = \sum_i w_{k i}\, f(a_i)
$$

$$
\frac{\partial a_k}{\partial a_i}
= \frac{\partial}{\partial a_i} \left( \sum_i w_{k i}\, f(a_i) \right)
= w_{k i}\, f'(a_i)
$$


Now we have everything we need for the weight-update rule

$$
\boxed{
\frac{\partial a_k}{\partial a_i}
= w_{k i}\, f'(a_i)
}
$$

$$
\boxed{
  \frac{\partial}{\partial a_i}
  \frac{\partial E}{\partial x_j}
  = \sum_{k}\left[
\left(
\frac{\partial}{\partial a_i}
\frac{\partial E}{\partial x_k}
\right)
\left(f'(a_k)\,w_{k j}\right)
\;+\;
\left(\frac{\partial E}{\partial x_k}\right)
\left(
f''(a_k)\,\frac{\partial a_k}{\partial a_i}\,w_{k j}
\right)
\right]
}
$$

$$
\boxed{
\frac{\partial^2 E}{\partial a_i^2} =
\left(
  \frac{\partial}{\partial a_i}
  \frac{\partial E}{\partial x_i}
\right)
f'(a_i)
\;+\;
\left(
  \frac{\partial E}{\partial x_i}
\right)
f''(a_i)
}
$$

$$
\boxed{
\frac{\partial^2 E}{\partial w_{ij}^2}
= \frac{\partial^2 E}{\partial a_i^2} \, x_j^2
}
$$


## Showing equivalence to paper

Starting from
$$
\frac{\partial^2 E}{\partial a_i^2}
=
\left(
\frac{\partial}{\partial a_i}
\frac{\partial E}{\partial x_i}
\right)
f'(a_i)
+
\left(
\frac{\partial E}{\partial x_i}
\right)
f''(a_i)
$$

Use the recursion
$$
\frac{\partial}{\partial a_i}\frac{\partial E}{\partial x_j}
=
\sum_k
\left[
\left(
\frac{\partial}{\partial a_i}
\frac{\partial E}{\partial x_k}
\right)
f'(a_k)w_{k j}
+
\left(
\frac{\partial E}{\partial x_k}
\right)
f''(a_k)
\frac{\partial a_k}{\partial a_i}
w_{k j}
\right]
$$

Set $j=i$:
$$
\frac{\partial}{\partial a_i}\frac{\partial E}{\partial x_i}
=
\sum_k
\left[
\left(
\frac{\partial}{\partial a_i}
\frac{\partial E}{\partial x_k}
\right)
f'(a_k)w_{k i}
+
\left(
\frac{\partial E}{\partial x_k}
\right)
f''(a_k)
\frac{\partial a_k}{\partial a_i}
w_{k i}
\right]
$$

Product rule identity
$$
\frac{\partial}{\partial a_i}
\left(\frac{\partial E}{\partial a_k}\right)
=
\left(
\frac{\partial}{\partial a_i}
\frac{\partial E}{\partial x_k}
\right)
f'(a_k)
+
\left(
\frac{\partial E}{\partial x_k}
\right)
f''(a_k)
\frac{\partial a_k}{\partial a_i}
$$

So
$$
\frac{\partial}{\partial a_i}\frac{\partial E}{\partial x_i}
=
\sum_k
w_{k i}
\frac{\partial}{\partial a_i}
\left(
\frac{\partial E}{\partial a_k}
\right)
$$

Local dependence
$$
\frac{\partial a_k}{\partial a_i}
=
w_{k i}f'(a_i)
$$

Thus (diagonal term)
$$
\frac{\partial}{\partial a_i}
\left(
\frac{\partial E}{\partial a_k}
\right)
\approx
\frac{\partial^2 E}{\partial a_k^2}
w_{k i}
f'(a_i)
$$

Plug in:
$$
\frac{\partial}{\partial a_i}
\frac{\partial E}{\partial x_i}
\approx
f'(a_i)
\sum_k
w_{k i}^2
\frac{\partial^2 E}{\partial a_k^2}
$$

Final expression
$$
\boxed{
\frac{\partial^2 E}{\partial a_i^2}
\approx
f'(a_i)^2
\sum_k
w_{k i}^2
\frac{\partial^2 E}{\partial a_k^2}
+
f''(a_i)
\frac{\partial E}{\partial x_i}
}
$$

Weight Hessian
$$
\boxed{
\frac{\partial^2 E}{\partial w_{i j}^2}
=
\frac{\partial^2 E}{\partial a_i^2}\,x_j^2
}
$$
